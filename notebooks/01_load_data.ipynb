# STEP 1 · 데이터 로드 


>  **CSV가 아니라 SAV를 쓴다.** 2025 CSV 파일은 인코딩이 깨져 있어
> 읽다가 에러가 난다. SAV(SPSS 원본)는 멀쩡하고 변수 라벨까지 들어있다.

```python
# pandas: 표 데이터를 다루는 핵심 도구
import pandas as pd
# pyreadstat: SPSS .sav 파일을 읽는 도구
import pyreadstat
```

```python
# 원본 데이터를 읽는다. meta에는 변수 설명(라벨)이 함께 들어온다.
df23, meta23 = pyreadstat.read_sav("../data/raw/journalist_2023.sav")
df25, meta25 = pyreadstat.read_sav("../data/raw/journalist_2025.sav")
```

### 1) 크기 확인 — 응답자 × 문항

```python
print("2023:", df23.shape)   # 기대값 (2011, 360)
print("2025:", df25.shape)   # 기대값 (2020, 271)
```

컬럼이 2023은 360개, 2025는 271개다.
그래서 STEP 2에서 '변수 번역'이 필요하다.

### 2) 결측치 확인 — 빈 칸이 있는가

```python
# 핵심 변수들의 빈 칸 개수를 센다.
key23 = ['q1_1','q6','q24_1','q26_3','q45_5','SQ1','DQ2']
key25 = ['q1_1','q5','q19_1','q21_3','q39_6','q38','SQ1','DQ2']
print("2023 결측:", {k:int(df23[k].isna().sum()) for k in key23})
print("2025 결측:", {k:int(df25[k].isna().sum()) for k in key25})
```

**결과: 전부 0이다.** 핵심 변수에 빈 칸이 하나도 없다.

### 3) 값 범위 점검 — 척도가 정상인가

```python
# 5점 척도는 1~5만, 10점 척도는 0~10만 있어야 정상이다.
print("2023 q1_1(5점):", sorted(df23['q1_1'].dropna().unique().astype(int)))
print("2025 q5(5점):  ", sorted(df25['q5'].dropna().unique().astype(int)))
print("2025 q19_1(10점):", sorted(df25['q19_1'].dropna().unique().astype(int)))
```

**결과: 정상.** 5점은 1~5, 10점은 0~10만 들어있다.
'99(무응답 코드)' 같은 이상값이 없다.
