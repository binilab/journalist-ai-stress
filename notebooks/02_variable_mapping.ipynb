# STEP 2 · 변수 매핑 
**같은 걸 묻는데 2023과 2025의 변수 이름이 다르다.**

| 개념 | 2023 | 2025 |
|---|---|---|
| 디지털 피로도 | `q6` | `q5` |
| 직업 만족도 | `q24_1` | `q19_1` |
| 번아웃-탈진 | `q26_3` | `q21_3` |


**변수 번역**을 코드 파일 하나에 남겨둔다 (`src/variable_mapping.py`).

```python
import sys; sys.path.insert(0, '..')
import pyreadstat, pandas as pd
from src.variable_mapping import MAPPING, get_var, MEDIA_LABELS

df23, _ = pyreadstat.read_sav("../data/raw/journalist_2023.sav")
df25, _ = pyreadstat.read_sav("../data/raw/journalist_2025.sav")
```

### 변수 번역이 작동하는지 테스트

```python
# 개념 이름만 주면 연도별 실제 변수명을 돌려준다.
for concept in ['digital_fatigue','satis_job','ai_jobthreat','ai_misinfo']:
    print(f"{concept:16s} 2023:{str(get_var(concept,2023)):8s} 2025:{get_var(concept,2025)}")
```

**눈여겨볼 점: `ai_misinfo`의 2023이 `None`이다.**
'검증 안 된 자료 기반 오보' 문항은 2025에 **새로 생겼다.**
이건 단순 누락이 아니라 **분석 무기**다 — 2023엔 없던 걱정이
2025에 생겼다는 건, AI 오보 우려가 그만큼 구체화됐다는 증거다.

### 공통 변수가 몇 개인지 확인

```python
common = sorted(set(df23.columns) & set(df25.columns))
print("두 해 공통 변수:", len(common), "개")
# 이 중 우리가 분석에 쓰는 건 인식(q1), 역할(q2,q3), 인구통계 등
```

### 매핑표를 CSV로 저장 (산출물)

```python
rows = []
for concept, years in MAPPING.items():
    rows.append({'개념': concept, '2023': years[2023], '2025': years[2025]})
vmap = pd.DataFrame(rows)
vmap.to_csv("../outputs/tables/variable_map.csv", index=False, encoding='utf-8-sig')
print("저장 완료: outputs/tables/variable_map.csv")
vmap
```
