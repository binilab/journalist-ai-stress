# STEP 8 · 정책 제안 정리

분석 결과를 정책으로 연결하는 단계. 본문은 `docs/policy_proposal.md`에 있다.

**핵심 원칙: 모든 제안은 [발견]->[제안]으로 근거를 단다.**
분석 결과에 근거한 정책 제안을 정리한다.

| 발견 | 제안 |
|---|---|
| AI 가이드라인 인지율 43% | ① 가이드라인 보급·교육 허브 |
| 고위험 2집단(과부하형·소외형) | ② 위기집단 2-트랙 맞춤 지원 |
| 피로 +7.2%p, 번아웃 역설 | ③ 디지털 피로 정기 모니터링 |

주최가 KPF이므로 제안의 주어는 '재단'으로 맺는다.
